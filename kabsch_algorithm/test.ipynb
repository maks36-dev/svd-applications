```python
import numpy as np
import matplotlib.pyplot as plt
from kabsch import kabsch_algorithm
```

```python
def dist(A, B):
    dist = 0
    for i in range(len(A)):
        dist += np.sum((A[i] - B[i])**2)**0.5
    return dist
```

```python
n = 5

graph = np.random.rand(n, n)
graph[graph > 0.5] = 1
graph[graph <= 0.5] = 0
```

```python
poinst = np.random.rand(n, 2)
poinst2 = np.random.rand(n, 2)
# poinst2 = poinst + np.array([1,2])
# poinst, poinst2
```

```python
def plot_graph(poinst, graph):
    plt.scatter(poinst[:, 0], poinst[:, 1])
    # for i in range(len(graph)):
    #     for j in range(len(graph)):
    #         if graph[i,j]:
    #             plt.plot([poinst[i][0], poinst[j][0]], [poinst[i][1], poinst[j][1]])
    plt.show()
plot_graph(poinst, graph)
```

```python
A = poinst
B = poinst2

H = kabsch_algorithm(A, B)
print(dist(H, B))
plt.plot(A[:, 0], A[:, 1], linewidth=2, c='b', label="1-ая молекула")
plt.scatter(A[:, 0], A[:, 1])
plt.plot(B[:, 0], B[:, 1], linewidth=2, c='r', label="2-ая молекула")
plt.scatter(B[:, 0], B[:, 1])
plt.plot(H[:, 0], H[:, 1], linewidth=2, linestyle=':', marker='o', c='g', label="Результат вращения")

plt.legend()

plt.xlabel('X-ось')
plt.ylabel('Y-ось')
plt.title('Алгоритм Кабша')
```

```python
phi = 3.14/6
S = np.array([
    [np.cos(phi), np.sin(phi)],
    [np.sin(phi), -np.cos(phi)]
])

A = poinst
B = poinst2

H = A.dot(S)
print(dist(H, B))
plt.plot(A[:, 0], A[:, 1], linewidth=2, c='b', label="1-ая молекула")
plt.scatter(A[:, 0], A[:, 1])
plt.plot(B[:, 0], B[:, 1], linewidth=2, c='r', label="2-ая молекула")
plt.scatter(B[:, 0], B[:, 1])
plt.plot(H[:, 0], H[:, 1], linewidth=2, linestyle=':', marker='o', c='g', label="Результат вращения")

plt.legend()

plt.xlabel('X-ось')
plt.ylabel('Y-ось')
plt.title('Алгоритм Кабша')
```
